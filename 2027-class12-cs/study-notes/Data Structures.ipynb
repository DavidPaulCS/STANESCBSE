```python
Stack:
    1. Stack is linear list that follows LIFO (Last In First Out)
        technique.
    2. Insertion & Deletion of an element in the stack is done only
        in one end (top end).
    3. The element which is inserted last (top element) 
        will deleted first.
    
Stack Operations:
    1. PUSH(x)      - inserts the element x in the stack at the end.
    2. POP()        - deletes the last (top) element.
    3. PEEK()/TOP() - view the last (top) element of the stack.
    4. TRAVERSE()   - process the elements of the stack.
    
Stack Applications:
    1. Browsing History
    2. Back button in the explorer
    3. Undo operation, etc.
```

```python
print("STACK OPERATIONS")
print("1.PUSH")
print("2.POP")
print("3.PEEK")
print("4.DISPLAY")
print("5.EXIT")

def PUSH(elem):
    if len(stack)>limit:
        print("STACK OVERFLOW")
    else:
        stack.append(elem)
        print(f"The element {elem} is PUSHED into the stack")
        print("STACK CONTAINS: ",stack)

def POP():
    if stack==[]:
        print("STACK UNDERFLOW")
    else:
        delelm=stack.pop()
        print(f"The top element {delelm} is deleted")
        print("STACK CONTAINS: ",stack)

def PEEK():
    if stack==[]:
        print("STACK UNDERFLOW")
    else:
        print(f"The Top Element in the stack is {stack[-1]}")

def DISPLAY():
    if stack==[]:
        print("STACK UNDERFLOW")
    else:
        for loop in range(len(stack)-1,-1,-1):
            print(stack[loop])
            
stack=[]
limit=10
ch="Y"
while ch=="Y" or ch=="y":
    choice=int(input("Enter your choice: "))
    if choice==1:
        video=input("Watched Video: ")
        PUSH(video)
    elif choice==2:
        POP()
    elif choice==3:
        PEEK()
    elif choice==4:
        DISPLAY()
    elif choice==5:
        pass
    else:
        print("INVALID CHOICE... Please choose from 1-4")
    ch=input("Do you want to continue: ")
```
