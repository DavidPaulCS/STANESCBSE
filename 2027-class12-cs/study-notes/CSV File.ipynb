```python
import csv
emp=[[1,"Aakash","HR"],[2,"Harshith","AC"],[3,"Kailash","Manager"]]
with open("Employee.csv","w+",newline='') as cf:
    wr=csv.writer(cf)
    wr.writerow(['EMPNO','EMPNAME',"DESIGNATION"])
    for ls in emp:
        wr.writerow(ls)
    cf.seek(0)
    readdata=csv.reader(cf)
    for elem in readdata:
        print(elem)
```

```python
import csv
emp=[[1,"Aakash","HR"],[2,"Harshith","AC"],[3,"Kailash","Manager"]]

with open("Employee.csv","w+",newline='') as cf:
    wr=csv.writer(cf)
    wr.writerow(['EMPNO','EMPNAME',"DESIGNATION"])
    wr.writerows(emp)
    cf.seek(0)
    readdata=csv.reader(cf)
    for elem in readdata:
        print(elem)
```

```python
NOTE: writerows() is used to write multiple rows.
      It accepts only an iterable (list of lists) or (list of tuples)
      (tuple of lists) or (tuple of tuples)
```
